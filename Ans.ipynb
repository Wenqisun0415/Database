## Q8

CAP theorem says that we can only have two out of consistency, partition tolerance and availability.

Say, we now have server sites A, B and C. Now, server site C lost its network connection. Any new updates happen in server site C will not be shown in A and B. If we want server site A and B to have data consistency, we will need to shut down all our servers(since the data in A is different with C, we don't want our user to see different results in A and C, the only thing we can do is to shut down all of them). If we choose availability, our user can still have the service, but they will have consistency issue.

Which way to go depends on the business rule and actual situation, if we are running a bank, we would want consistency. If we are running a social media, we would want availability.

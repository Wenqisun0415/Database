## Q8

CAP theorem says that we can only have two out of consistency, partition tolerance and availability.

Say, we now have server sites A, B and C. Now, server site C lost its network connection. Any new updates happen in server site C will not be shown in A and B. If we want server site A and B to have data consistency, we will need to shut down all our servers(since the data in A is different with C, we don't want our user to see different results in A and C, the only thing we can do is to shut down all of them). If we choose availability, our user can still have the service, but they will have consistency issue.

Which way to go depends on the business rule and actual situation, if we are running a bank, we would want consistency. If we are running a social media, we would want availability.

## Q9

### A)

* OR(Object Relational) impedance mismatch. Relational database are not designed for storing object. When we have an object like the invoice in Week6, we need to split it into several tables before we can put it into the relational database. When we want to retrieve it, we need to join those tables. It is a waste of effort. But NoSQL are designed to store objects. We can get rid of those normalization and denormalization process.

* To handle large data volumes and throughput. For a relational database, join several tables together will be expensive, the larger the data volume, the more number of tables we will need to join. However, there is no such join issue with NoSQL.

### B)

In relational database, the table, attributes and data type for each attributes together form the schema. It prevents users from putting bad data into database. Say, if the table only got $4$ columns, you want to put a row of $5$ elements into the database, it won't allow you do that. But for NoSQL, there is no schema. For MangoDB, you can put an object with $5$ attributes in it. After that, you can still put an object with only $1$ attribute into the database. It won't check whether you data is valid or not. The job is left with the user. The database can then focus on storing things, which results in a great improve in speed.
